```python
from pytrends.request import TrendReq
import pandas as pd
import plotly.graph_objects as go
```

```python
hl = 'ja-jp'
tz = -540
geo = "JP"
```

```python
kw_list = ['フィンテック', 'インシュアテック', 'サスティナブル', 'DeFi', 'STO']
timeframe = '2020-1-1 2021-12-31'
```

```python
pytrend = TrendReq(hl=hl, tz=tz)
pytrend.build_payload(kw_list=kw_list, timeframe=timeframe, geo=geo)
interest_over_time_df = pytrend.interest_over_time()
interest_over_time_df.head(n=100)
```

```python
df = pytrend.interest_over_time()
df.to_csv("./brand.csv")
df = pd.read_csv("./brand.csv", engine="python")
go.Figure(
    [
        go.Scatter(x=df["date"], y=df["フィンテック"], name="フィンテック"),
        go.Scatter(x=df["date"], y=df["インシュアテック"], name="インシュアテック"),
        go.Scatter(x=df["date"], y=df["サスティナブル"], name="サスティナブル"),
        go.Scatter(x=df["date"], y=df["DeFi"], name="DeFi"),
        go.Scatter(x=df["date"], y=df["STO"], name="STO"),
    ]
)
```

```python
kw_list = ['フィンテック']
timeframe = '2020-2-14 2021-2-21'
```

```python
pytrend.build_payload(kw_list=kw_list, timeframe=timeframe, geo=geo)
interest_by_region_df = pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=True)
interest_by_region_df.sort_values('geoCode').head(n=100)
```

```python
interest_by_region_df.reset_index().plot(x='geoName', y='フィンテック', figsize=(120, 10), kind ='bar')
```
